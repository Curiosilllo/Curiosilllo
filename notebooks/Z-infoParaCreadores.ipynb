## Instalación de Sphinx y generación de la documentación

Aquí se muestran algunos paquetes que necesitan instalar los creadores del curso para crear la documentación, etc.

+ Instalamos ```Sphinx```
```bash
    > conda install -c conda-forge sphinx
```


+ Instalamos ```nbsphinx```
```bash
    > conda install -c conda-forge nbsphinx
```


A partir de aquí probablemente no haya que hacerlo más porque vendrá del repositorio:

+ Seguimos el [manual de nbsphynx](https://nbsphinx.readthedocs.io/en/0.2.15/usage.html):
```bash
    > sphinx-quickstart
```
**Opciones que elegimos:** Separar directorio fuente y compilado: **NO**, Lenguaje del proyecto: **es**

+ Editamos ```source/conf.py```:
```
extensions = [
    'nbsphinx',
    'sphinx.ext.mathjax',
]
exclude_patterns = ['_build', 'Thumbs.db', '.DS_Store', '**.ipynb_checkpoints']
```


+ Editamos ```source/index.rst``` para cambiar e título de la documentación y alguna otra cosa que sale en inglés así como para añadir los ficheros que deben usarse en la misma (en principio todos los notebooks).

+ Para compilar la documentación (se ha actualizado el makefile para borrar ficheros que no deben subirse al repo y otras cosillas):
```bash
    > make html
```


+ Para borrar toda la documentación (para regenerarla desde 0, por si hay cambios grandes o ficheros que cambian de nombre, etc):
```bash
    > make clean
```


## Uso de Git y GitHub

[Tutorial básico de Git](https://www.diegocmartin.com/tutorial-git/)


+ **¡Muy importante!** Una vez que nos traigamos el repositorio clonado debemos actualizar nuestro fichero ```.git/config``` y añadir estas líneas para que al hacer los commits a los notebooks se les limpien automáticamente las salidas:
```
[filter "strip-notebook-output"]
    clean = "jupyter nbconvert --ClearOutputPreprocessor.enabled=True --to=notebook --stdin --stdout --log-level=ERROR"
```


+ Clonar el repositorio en nuestra máquina local:
```bash
    > git clone https://github.com/zerjillo/cursoAstronomia.git
```


+ Crear un ```commit``` (añadiendo archivos nuevos):
```bash
    > git add .
    > git commit -a -m "Mensaje del commit"
```


+ Actualizar la copia local con la copia en GitHub:
```bash
    > git pull
```


+ Mandando los commits locales a la copia en GitHub:
```bash
    > git push
```


+ Comprobar los cambios que faltan por aplicar (ficheros cambiados, añadidos, borrados)
```bash
    > git status
```
