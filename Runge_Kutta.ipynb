$Runge-Kutta$ $de$ $Segundo$ $orden$



```python
def funci(y,t):
  return -0.4*y+0.2

def RGK2(a,b,h,f,alpha):
  y=alpha
  t=a
  #condicionante
  FLAG=1
  #Acumulador
  tgraf=[]
  ygraf=[]
  while (FLAG==1):
    tgraf.append(t)
    ygraf.append(y)
    k1=h*f(y,t)
    k2=h*f(y+k1,t+h)
    t=t+h
    y=y+0.5*(k1+k2)
    if t>b:
      FLAG=0
      return [tgraf,ygraf]
  
  
```

```python
import matplotlib.pyplot as plt

a=RGK2(0,10,0.1,funci,0)[0]
b=RGK2(0,10,0.1,funci,0)[1]

plt.scatter(a, b)
```

```python
def RK4_EDO_2Grado(F,h,n,X0,t0):
  #n es el núm de pasos
  X=[None]*n #a parte de ser un vector cada elemento es un vector
  #X va a ser un vector de n elementos, vacíos
  X[0]=np.array(X0)
  #primer elemento del vector
  global T
  T=[t0+i*h for i in range(n)]
  for i in range(n-1):
    K1=h*F(T[i],X[i])
    K2=h*F(T[i]+h/2,X[i]+K1/2)
    K3=h*F(T[i]+h/2,X[i]+K1/2)
    K4=h*F(T[i]+h/2,X[i]+K3)
    X[i+1]=X[i]+(1/6)*(K1+2*K2+2*K3+K4)
  return X
```

$Atractor$ $de$ $Lorentz$ $con$ $método$ $de$ $Euler$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
a=10
b=28
c=8/3
def lorenz(x,y,z):
  u= a*(y-x)
  v=x*(b-z)-y
  w=x*y-c*z
  return u,v,w
dt=0.01
stepCnt=10000

xs=np.empty((stepCnt +1,))
ys=np.empty((stepCnt +1,))
zs=np.empty((stepCnt +1,))

xs[0], ys[0], zs[0]= (-1.3560, -2.492152,12.317410)

for i in range(stepCnt):
  x_dot,y_dot,z_dot= lorenz(xs[i],ys[i],zs[i])
  #Usamos el método de euler
  xs[i+1]=xs[i]+(x_dot*dt)
  ys[i+1]=ys[i]+(y_dot*dt)
  zs[i+1]=zs[i]+(z_dot*dt) 

fig=plt.figure(figsize=(12,12))
ax=fig.gca(projection="3d")

ax.plot(xs,ys,zs)
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_zlabel("Eje Z")
ax.set_title("Atractor de Lorenz")
```

```python
fig, ax= plt.subplots()
ax.set(xlabel="Eje Y",ylabel="Eje Z", title="Atractor de lorenz")
plt.plot(xs,zs,":",label="Z vs Y")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
#Ecuación diferencial a resolver
def f(t,x):
  a=10.0
  b=28.0
  c=8/3.0
  return np.array([a*(x[1]-x[0]),
                  x[0]*(b-x[2])-x[1],
                  x[0]*x[1]-c*x[2]])
#Condiciones iniciales
t0=0
#x0=np.array([3,0,0])
x0=np.array([-1.3560, -2.492152,12.317410])
#Intervalo solución y tamaño de paso
tn=200
h=0.01
#Método numérico
n=round((tn-t0)/h)
t=np.linspace(t0,tn,n+1)
x=np.zeros((n+1,len(x0)))
x[0,:]=x0

for i in range(n):
  k1=h*f(t[i],x[i,:])
  k2=h*f(t[i]+h/2,x[i,:]+k1/2)
  k3=h*f(t[i]+h/2,x[i,:]+k2/2)
  k4=h*f(t[i]+h,x[i,:]+k3)
  x[i+1,:]=x[i,:]+(1/6)*(k1+k2*2+k3*2+k4)
#Gráfica de la solución
fig=plt.figure(figsize=(12,12))
ax=plt.axes(projection="3d")
ax.plot3D(x[:,0],x[:,1],x[:,2])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import numba
import time

start_time = time.clock()
 
w=1
b=0.1
@numba.jit()
def V(u,t):
    x1,dx1, x2, dx2=u
    ddx1=-w**2 * x1 -b * dx1
    ddx2=-(w+0.5)**2 * x2 -(b+0.1) * dx2
    return np.array([dx1,ddx1,dx2,ddx2])


@numba.jit()
def rk4(f, u0, t0, tf , n):
    t = np.linspace(t0, tf, n+1)
    u = np.array((n+1)*[u0])
    h = t[1]-t[0]
    for i in range(n):
        k1 = h * f(u[i], t[i])    
        k2 = h * f(u[i] + 0.5 * k1, t[i] + 0.5*h)
        k3 = h * f(u[i] + 0.5 * k2, t[i] + 0.5*h)
        k4 = h * f(u[i] + k3, t[i] + h)
        u[i+1] = u[i] + (k1 + 2*(k2 + k3) + k4) / 6
    return u, t

u, t  = rk4(V,np.array([0,0.2,0,0.3]) ,0,100, 20000)

print("Execution time:",time.clock() - start_time, "seconds")
x1,dx1,x2,dx2 = u.T
plt.plot(x1,x2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
